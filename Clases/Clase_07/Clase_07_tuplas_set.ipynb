## Parte 1. Tuplas: Inmutabilidad, Creación y Acceso

### 1.1 Definición

Una **tupla** es una estructura de datos en Python que permite almacenar una colección **ordenada** e **inmutable** de elementos. A diferencia de las listas, una vez creada una tupla, no se pueden modificar sus elementos (no se pueden agregar, eliminar ni cambiar).

**Características principales:**
- **Inmutables**: No se pueden modificar después de su creación
- **Ordenadas**: Mantienen el orden de inserción
- **Permiten duplicados**: Pueden contener elementos repetidos
- **Heterogéneas**: Pueden almacenar diferentes tipos de datos
- **Indexables**: Se accede a sus elementos mediante índices (desde 0)

### 1.2 Sintaxis

````python
# Creación con paréntesis
tupla = ("elemento1", "elemento2", "elemento3")
````

````python
# Creación sin paréntesis (empaquetado)
tupla = "elemento1", "elemento2", "elemento3"
````

````python
# Tupla vacía
tupla_vacia = ()
````

````python
# Tupla de un solo elemento (requiere coma)
tupla_unitaria = ("elemento",)
````

````python
# Creación con el constructor tuple()
tupla = tuple([1, 2, 3])
````

````python
# Acceso por índice
elemento = tupla[0]  # Primer elemento
elemento = tupla[-1]  # Último elemento
````

````python
# Slicing (rebanado)
sub_tupla = tupla[0:2:1]

print(tupla)
````

## Parte 2. Conjuntos (set): Eliminación de Duplicados y Operaciones Básicas

### 2.1 Definición

Un **conjunto (set)** es una estructura de datos en Python que almacena una colección **no ordenada** de elementos **únicos**. Los conjuntos son ideales para eliminar duplicados y realizar operaciones matemáticas como unión, intersección y diferencia.

**Características principales:**
- **Sin duplicados**: Cada elemento aparece solo una vez
- **No ordenados**: No mantienen orden de inserción
- **Mutables**: Se pueden agregar o eliminar elementos
- **No indexables**: No se puede acceder por índice
- **Elementos inmutables**: Solo pueden contener tipos de datos inmutables (int, str, tuple, etc.)

### 2.2 Sintaxis

````python
# Creación con llaves
conjunto = {1, 2, 3}
````

````python
# Conjunto vacío (debe usar constructor)
conjunto_vacio = set()
````

````python
# Creación desde una lista (elimina duplicados)
conjunto = set([1, 2, 2, 3, 3, 3])  # Resultado: {1, 2, 3}
print(f"Conjunto desde lista: {conjunto}")
````

````python
# Agregar elementos
conjunto.add(4)
print(f"Después de agregar 4: {conjunto}")
````

````python

# Eliminar elementos
conjunto.discard(2)  # No genera error si no existe
print(f"Después de eliminar 2: {conjunto}")
````

````python
# Operaciones entre conjuntos
conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}

union = conjunto1 | conjunto2
interseccion = conjunto1 & conjunto2
diferencia = conjunto1 - conjunto2
diferencia_simetrica = conjunto1 ^ conjunto2

print(f"Unión: {union}")
print(f"Intersección: {interseccion}")
print(f"Diferencia: {diferencia}")
print(f"Diferencia simétrica: {diferencia_simetrica}")
````

### 2.3 Ejemplos Básicos

**Ejemplo 1: Eliminar números duplicados**

````python
# Lista con números duplicados
numeros = [1, 2, 2, 3, 4, 4, 4, 5, 1, 3]

print(f"Lista original: {numeros}")
print(f"Total de elementos: {len(numeros)}")

# Convertir a conjunto para eliminar duplicados
numeros_unicos = set(numeros)

print(f"Números únicos: {numeros_unicos}")
print(f"Total de únicos: {len(numeros_unicos)}")

# Convertir de nuevo a lista ordenada
lista_limpia = sorted(numeros_unicos)
print(f"Lista limpia y ordenada: {lista_limpia}")
````

**Ejemplo 2: Operaciones entre conjuntos de estudiantes**

````python
# Estudiantes inscritos en diferentes cursos
python = {"Ana", "Luis", "María", "Carlos", "Elena"}
javascript = {"María", "Pedro", "Carlos", "Sofía"}

print(f"Estudiantes de Python: {python}")
print(f"Estudiantes de JavaScript: {javascript}")

# Estudiantes en ambos cursos (intersección)
ambos_cursos = python & javascript
print(f"Estudiantes en ambos cursos: {ambos_cursos}")

# Todos los estudiantes (unión)
todos = python | javascript
print(f"Total de estudiantes: {todos}")

# Solo en Python (diferencia)
solo_python = python - javascript
print(f"Solo en Python: {solo_python}")

# En solo uno de los cursos (diferencia simétrica)
solo_uno = python ^ javascript
print(f"En solo un curso: {solo_uno}")
````

**Ejemplo 3: Gestión dinámica de un conjunto**

````python
# Conjunto de frutas favoritas
frutas = {"manzana", "banana", "naranja"}
print(f"Frutas iniciales: {frutas}")

# Agregar una fruta
frutas.add("uva")
print(f"Después de agregar uva: {frutas}")

# Intentar agregar duplicado (no tendrá efecto)
frutas.add("manzana")
print(f"Después de intentar agregar manzana duplicada: {frutas}")

# Eliminar una fruta
frutas.discard("banana")
print(f"Después de eliminar banana: {frutas}")

# Verificar si una fruta está en el conjunto
if "naranja" in frutas:
    print("La naranja está en el conjunto")

# Número de elementos
print(f"Total de frutas: {len(frutas)}")
````
